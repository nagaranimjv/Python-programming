Problem 4 – Using replace()
Topics: String method replace


```python
# Write a program that:
# 1. Has a string:
#       sentence = "I like Java programming"
# 2. Use replace() to change "Java" to "Python".
# 3. Print the original string and the new string.

# Observe that the original string does not change unless you store the result.
```

Problem 5 – Using find, count, index, strip

```python
# Write a program that:
# 1. Has a string:
#       text = "   banana mango banana   "
# 2. Use strip() to remove spaces from both ends and print the result.
# 3. Use count() to count how many times "banana" appears.
# 4. Use find() to find the position of the first "mango".
# 5. Use index() to find the position of the first "banana".

### (Hint: index() will give error if substring not found – here it will be found.)
```

```python
# 4. Use find() to find the position of the first "mango".
print(text.find("mango"))
```

```python
# 5. Use index() to find the position of the first "banana".
print(text.index("banana"))
```

Problem 7 – split and join
Topics: split, join

```python
# Write a program that:
# 1. Has a string:
#       fruits = "apple,banana,orange"
# 2. Use split(",") to break it into separate items and print the result.
# 3. Then join them back into a string using " - " as separator, like:
#       "apple - banana - orange"
#   and print this new string.

#(Hint: use " - ".join(<list_from_split>))
```

```python
# 1. Has a string:
#       fruits = "apple,banana,orange"
fruits = "apple,banana,orange"
print(fruits)
```

```python
# 2. Use split(",") to break it into separate items and print the result.
fruits_split = (fruits.split(","))
fruits_split
```

```python
# 3. Then join them back into a string using " - " as separator, like:
fruits_join = "-".join(fruits_split)
fruits_join
```

Problem 9 – Email Username Extractor
Topics: find, slicing

```python
# Write a program that:
# 1. Asks the user to enter an email id, like "student123@gmail.com".
# 2. Use find("@") to locate the position of '@'.
# 3. Use slicing to extract the part before '@' as the username.
# 4. Print:
#       "Your username is: <username>"
```

```python
# 1. Asks the user to enter an email id, like "student123@gmail.com".
mail_id  = input("Enter mail id:")
print(mail_id)
```

```python
# 2. Use find("@") to locate the position of '@'.
print(mail_id.find("@"))
```

```python
# 3. Use slicing to extract the part before '@' as the username.
position = mail_id.index("@")
name = mail_id[:position]
print(name)
print(position)
```
